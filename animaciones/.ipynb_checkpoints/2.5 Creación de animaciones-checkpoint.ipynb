### 1.2 Animaciones para coordenadas cartesianas

Se generará una animacion con el uso de la funcion SENO y se propondrán parametros que modifiquen el comportamiento del grafico.

```python
# 1- importamos librerias
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation, rc
from IPython.display import HTML
```

```python
# 2 - tamaño del grafico
fig, ax = plt.subplots()
ax.set_xlim(( 0, 4))
ax.set_ylim((-2, 2))
line, = ax.plot([], [], 'r-', lw=4)
```

```python
# 3- funciones de inicio / actualización

def init():
    line.set_data([], [])
    return (line,)


def animate(i):
    x = np.linspace(0, 4, 1000)
    y = i*0.01*np.sin(np.pi*x)
    line.set_data(x, y)
    return (line,)
```

```python
# 4- animacion final
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=30, 
                               blit=True)
#anim.save('animation2.gif', writer="pillow", fps=60)
HTML(anim.to_jshtml())
```

### 1.3 Animaciones para coordenadas polares - Rosa Polar

Mediante matematicas se generaran los petalos de una rosa.

```python
# 2 - tamaño del grafico
fig2, ax2 = plt.subplots(subplot_kw={'projection': 'polar'})
ax2.set_rlim(0,4)
rosa, = ax2.plot([], [], 'y-', lw=2)
```

```python
# 3- funciones de inicio / actualización

def initRosa():
    rosa.set_data([], [])
    return (rosa,)


def animateRosa(i):
    petalos = 5
    theta = np.linspace(0, 2*np.pi, 500)
    r = (i*0.1)* np.sin(petalos * theta)
    rosa.set_data(theta, r)
    return (rosa,)
```

```python
# 4- animacion final
animRosa = animation.FuncAnimation(fig2, animateRosa, init_func=initRosa,
                               frames=40, interval=100, 
                               blit=True)
#anim.save('animation2.gif', writer="pillow", fps=60)
HTML(animRosa.to_jshtml())
```

### 1.4 Uso de SciPy en animaciones - interpolador cubico

Mediante el uso de la libreria SciPy se emulará el movimiento de una serpiente.

```python
#1 - importar librerias
from scipy.interpolate import interp1d
import random
```

```python
# 2 - tamaño del grafico
fig3, ax3 = plt.subplots()
ax3.set_xlim(( 0, 11))
ax3.set_ylim(( -10, 10))
snake, = ax3.plot([], [], 'go-', lw=6)
```

```python
# 3- funciones de inicio / actualización

def initSnake():
    snake.set_data([], [])
    return (snake,)

xn = []
yn = []

def animateSnake(i):
    xn.append(i)
    yn.append(random.randrange(-5, 5, 1))
    
    x = np.array(xn)
    y = np.array(yn)
    
    if(len(x) > 3):
        xnew = np.linspace(0, i, num=i*10, endpoint=True)
        f = interp1d(x, y)
        f2 = interp1d(x, y, kind='cubic')
        snake.set_data(xnew, f2(xnew))
        return (snake,)
    
    else :
        snake.set_data(xn, yn)
        return (snake,)
```

```python
# 4- animacion final
animSnake = animation.FuncAnimation(fig3, animateSnake, init_func=initSnake,
                               frames=10, interval=1000, 
                               blit=True)
#anim.save('animation2.gif', writer="pillow", fps=60)
HTML(animSnake.to_jshtml())
```
